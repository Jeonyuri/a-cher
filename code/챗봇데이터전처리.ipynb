```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
chunk_size = 500
num_chunks = 21
dfs = []

for i in range(num_chunks):
    start_index = i * chunk_size
    end_index = (i + 1) * chunk_size
    dfs.append(df[start_index:end_index])

# Access individual DataFrames like dfs[0], dfs[1], ..., dfs[9]
```

```python
for i in range(21):
  print("dfs[%d]['embedding'] = pd.Series([[]] * len(dfs[%d]))" %(i, i))
  print("dfs[%d]['embedding'] = dfs[%d]['유저'].map(lambda x: list(encoder.encode(x)))" %(i, i))
  print("dfs[%d].head()" %i)
```

```python
df_0 = df[:500]
```

```python
df_0['embedding'] = pd.Series([[]] * len(df_0))
df_0['embedding'] = df_0['유저'].map(lambda x: list(encoder.encode(x)))
df_0.head()
```

```python
df_0.to_excel('500question.xlsx')
```
